```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import shapiro
from scipy.stats import kstest
from scipy import stats
%matplotlib inline
plt.rcParams['figure.figsize'] = [16, 9]
plt.rcParams['figure.dpi'] = 40
plt.rcParams.update({'font.size': 18})
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
screen = df['Screen Size']
weight = df['Weight']
brightness = df['Brightness']
```

```python
class ProjectThree():
    def __init__(self, data):
        self.data = data
        self.data_chuan_hoa = np.array(data)
        self.data_chuan_hoa_1 = (self.data_chuan_hoa - np.mean(self.data_chuan_hoa))/np.std(self.data_chuan_hoa)
        
    def hist(self):
        sns.histplot(self.data, color='red', kde = True)

    def qq_plot(self):
        sm.qqplot(self.data_chuan_hoa_1, fit=True, line='45')
        
    def shapiro_test(self):
        print(f'Shapiro:')
        p_value_s = shapiro(self.data)[1]
        if p_value_s > 0.05:
            print('\tMẫu tuân theo phân phối chuẩn')
        else:
            print('\tMẫu không tuân theo phân phối chuẩn')
    
    def ks_test(self):
        print(f'Kolmogorov:')
        p_value_k =  kstest(rvs=self.data, cdf='norm', args=(np.mean(self.data), np.std(self.data)))[1]
        if p_value_k > 0.05:
            print('\tMẫu tuân theo phân phối chuẩn')
        else:
            print('\tMẫu không tuân theo phân phối chuẩn')
```

```python
screen_test = ProjectThree(screen)
weight_test = ProjectThree(weight)
brightness_test = ProjectThree(brightness)
```

```python
screen_test.hist()
screen_test.qq_plot()
```

```python
screen_test.shapiro_test()
screen_test.ks_test()
```

```python
weight_test.hist()
weight_test.qq_plot()
```

```python
weight_test.shapiro_test()
weight_test.ks_test()
```

```python
brightness_test.hist()
brightness_test.qq_plot()
```

```python
brightness_test.shapiro_test()
brightness_test.ks_test()
```

```python
pl = stats.levene(screen, weight, brightness)[1]
print(pl)
if pl > 0.05:
    print('Phương sai bằng nhau')
else:
    print('Có ít nhất một cặp Phương sai không bằng nhau')
```

```python
pb = stats.bartlett(screen, weight, brightness)[1]
print(pb)
if pb > 0.05:
    print('Phương sai bằng nhau')
else:
    print('Có ít nhất một cặp Phương sai không bằng nhau')
```
